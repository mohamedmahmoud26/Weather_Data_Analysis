```python
# Import libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create day data
days=np.arange(1,31) # Generate an array of days from 1 to 30

# Generate random weather data
humidity=np.random.randint(30, 90, size=30)   # Humidity between 30% and 90%
temperatures=np.random.randint(7,40,size=30) # Temperatures between 7°C and 40°C
wind_speed=np.random.randint(5,30,size=30) # Wind speed between 5 and 30 km/h
# This is 
# Create DataFrame
weather_data = pd.DataFrame({
    "Day": days,
    "Humidity(%)": humidity,
    "Temperature(°C)": temperatures,
    "Wind Speed(km/h)": wind_speed
})

# Display the data
weather_data
                           
```

```python
# Calculate average values
avg_humidity = weather_data["Humidity(%)"].mean()
avg_temperature = weather_data["Temperature(°C)"].mean()
avg_wind_speed = weather_data["Wind Speed(km/h)"].mean()

# Find the hot and cold day
max_temperature_day = weather_data.iloc[weather_data["Temperature(°C)"].idxmax()]
min_temperature_day = weather_data.iloc[weather_data["Temperature(°C)"].idxmin()]

# Print the results
print(f"Average Temperature: \n    {avg_temperature:.2f} °C")
print(f"Average Humidity: \n    {avg_humidity:.2f} %")
print(f"Average Wind Speed: \n    {avg_wind_speed:.2f} km/h")
print(f"Hot Day: \n    Day {max_temperature_day['Day']} ({max_temperature_day['Temperature(°C)']}°C)")
print(f"Cold Day:  \n    Day {min_temperature_day['Day']} ({min_temperature_day['Temperature(°C)']}°C)")
```

```python

# Set the figure size
plt.figure(figsize=(16,8))

# Subplot 1: Daily Temperature
plt.subplot(3, 1, 1)
plt.plot(weather_data['Day'], weather_data['Temperature(°C)'], 'r',marker='o', label="Temperature")
plt.title("Daily Temperature")
plt.xlabel("Day")
plt.xticks(weather_data['Day'])  # Display each day as a tick
# plt.yticks(weather_data['Temperature(°C)'])
plt.ylabel("Temperature (°C)")
plt.grid()  # Add grid lines
plt.legend()  # Add legend

# Subplot 2: Daily Humidity
plt.subplot(3, 1, 2)
plt.plot(weather_data['Day'], weather_data['Humidity(%)'], color='black',marker='o', label="Humidity")
plt.title('Daily Humidity')
plt.xlabel('Day')
plt.xticks(weather_data['Day'])  # Display each day as a tick
# plt.yticks(weather_data['Humidity(%)'])
plt.ylabel('Humidity (%)')
plt.grid()  # Add grid lines
plt.legend()  # Add legend

# Subplot 3: Daily Wind Speed
plt.subplot(3, 1, 3)
plt.plot(weather_data['Day'], weather_data['Wind Speed(km/h)'], 'b',marker='x', label='Wind Speed')
plt.title('Daily Wind Speed')
plt.xlabel('Day')
plt.xticks(weather_data['Day'])  # Display each day as a tick
# plt.yticks(weather_data['Wind Speed(km/h)'])
plt.ylabel('Wind Speed (km/h)')
plt.grid()  # Add grid lines
plt.legend()  # Add legend
# Display the plots
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
